```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

# Read in image

orgImage = mpimg.imread('test.jpg')
image = np.copy(orgImage)
print('This image is:', type(image), 'with dimensions:', image.shape)
```

```python
def showImage(img):
    plt.imshow(img)
    plt.show()
    
def showGrayImage(img):
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
image = np.copy(orgImage)
showImage(image)
```

```python
r = image[:, :, 0]
print(r.shape)
g = image[:, :, 1]
b = image[:, :, 2]
```

```python
rt = 210
gt = 210
bt = 210
r_sel = (r < rt)
r_sel.shape
```

```python
r_sel.dtype
```

```python
t_sel = (r < rt) | (g < gt) | (b < bt) # Note: '|' is the elementwise logical OR operation
t_sel.shape
```

```python
sel = image[t_sel]
type(sel)
```

```python
sel.shape
```

In general, when the boolean array has fewer dimensions than the array being indexed, this is equivalent to `y[b, ...]`, which means `y` is indexed by `b` followed by as many `:` as are needed to fill out the rank of `y`. Thus the shape of the result is one dimension containing the number of True elements of the boolean array, followed by the remaining dimensions of the array being indexed. Click [here](https://docs.scipy.org/doc/numpy/user/basics.indexing.html#boolean-or-mask-index-arrays) for more information.
